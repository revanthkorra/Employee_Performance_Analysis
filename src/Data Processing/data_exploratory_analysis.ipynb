### 4. Correlation Analysis

```python
# Correlation analysis for numerical variables
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_cols].corr()

# Plot correlation heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()

# Print correlations with Performance Rating
print("\nCorrelations with Performance Rating (sorted):")
correlations = correlation_matrix['PerformanceRating'].sort_values(ascending=False)
print(correlations)
```

###  Insights from Correlation Analysis

This correlation analysis helps us understand how various numerical features relate to **Performance Rating**:

####  Top Positively Correlated Features:
- **EmpEnvironmentSatisfaction** (`~0.40`):  
  Employees with higher environment satisfaction tend to perform better. This is one of the strongest signals in the dataset.
  
- **EmpLastSalaryHikePercent** (`~0.33`):  
  Higher recent salary hikes are moderately associated with better performance, suggesting reward-driven motivation.

- **EmpWorkLifeBalance** (`~0.12`):  
  Indicates a slight tendency that better work-life balance can contribute to better performance.

####  Features with Low or No Correlation:
- Variables like `DistanceFromHome`, `Age`, and `YearsSinceLastPromotion` show very low or no direct correlation with performance.
- These may still be useful in combination with other features but individually hold limited predictive power.

#### Conclusion:
- No variable shows a very strong (>0.6) correlation with performance — indicating that performance is likely influenced by **multiple moderate factors**, not a single dominant one.
- These insights help prioritize **feature selection** for modeling and highlight the importance of **multivariate analysis**.

### 8. Check for Outliers

```python
# Select numerical columns excluding binary/categorical variables and employee numbers
numerical_features = ['Age', 'DistanceFromHome', 'EmpEducationLevel', 'EmpEnvironmentSatisfaction',
                     'EmpHourlyRate', 'EmpJobInvolvement', 'EmpJobLevel', 'EmpJobSatisfaction',
                     'NumCompaniesWorked', 'EmpLastSalaryHikePercent', 'EmpRelationshipSatisfaction',
                     'TotalWorkExperienceInYears', 'TrainingTimesLastYear', 'EmpWorkLifeBalance',
                     'ExperienceYearsAtThisCompany', 'ExperienceYearsInCurrentRole',
                     'YearsSinceLastPromotion', 'YearsWithCurrManager']

# Create box plots for numerical features
plt.figure(figsize=(15, 10))
sns.boxplot(data=df[numerical_features])
plt.xticks(rotation=90)
plt.title('Box Plots of Numerical Features')
plt.tight_layout()
plt.show()

# Calculate outliers using IQR method
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    return len(outliers), lower_bound, upper_bound

# Print outlier statistics for each numerical feature
print("\nOutlier Analysis:")
print("\nFeature | Number of Outliers | Lower Bound | Upper Bound")
print("-" * 60)
for feature in numerical_features:
    n_outliers, lb, ub = identify_outliers(df, feature)
    print(f"{feature:30} | {n_outliers:5} | {lb:10.2f} | {ub:10.2f}")
```

```python
# Let's look at the distribution of features with significant outliers
features_with_outliers = ['NumCompaniesWorked', 'TotalWorkExperienceInYears', 
                         'TrainingTimesLastYear', 'ExperienceYearsAtThisCompany',
                         'YearsSinceLastPromotion']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_with_outliers, 1):
    plt.subplot(2, 3, i)
    sns.histplot(data=df, x=feature, bins=30)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()

# Calculate summary statistics for these features
print("\nSummary Statistics for Features with Significant Outliers:")
print(df[features_with_outliers].describe())
```

```python
# Let's examine if these outliers have any relationship with performance rating
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_with_outliers, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(data=df, x='PerformanceRating', y=feature)
    plt.title(f'{feature} by Performance Rating')
plt.tight_layout()
plt.show()

# Calculate mean performance rating for outlier vs non-outlier groups
print("\nPerformance Rating Analysis for Outliers vs Non-Outliers:")
print("\nFeature | Mean Rating (Non-Outliers) | Mean Rating (Outliers)")
print("-" * 65)

for feature in features_with_outliers:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outlier_mask = (df[feature] < lower_bound) | (df[feature] > upper_bound)
    mean_rating_outliers = df[outlier_mask]['PerformanceRating'].mean()
    mean_rating_non_outliers = df[~outlier_mask]['PerformanceRating'].mean()
    
    print(f"{feature:30} | {mean_rating_non_outliers:23.2f} | {mean_rating_outliers:19.2f}")
```

Based on the outlier analysis:

1. Features with significant outliers:
- TrainingTimesLastYear (188 outliers)
- YearsSinceLastPromotion (88 outliers)
- ExperienceYearsAtThisCompany (56 outliers)
- TotalWorkExperienceInYears (51 outliers)
- NumCompaniesWorked (39 outliers)

2. Impact on Performance Rating:
- Most outliers show slightly lower mean performance ratings
- YearsSinceLastPromotion outliers show the largest negative impact on performance (2.78 vs 2.96)
- TotalWorkExperienceInYears outliers show slightly higher performance (2.98 vs 2.95)

3. Nature of Outliers:
- Most outliers are on the upper end (higher values)
- They represent legitimate business cases (e.g., very experienced employees or long-term employees)
- No clear evidence of data errors

Given these findings, while there are outliers, they appear to be valid data points representing real business cases rather than errors. Therefore, removing them might not be appropriate for this analysis.

### Identify Top Factors affecting performance

```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from xgboost import XGBClassifier

# Separate features and target
X = df.drop('PerformanceRating', axis=1)
y = df['PerformanceRating']

# Remap y labels: 2 -> 0, 3 -> 1, 4 -> 2
y_mapped = y.replace({2: 0, 3: 1, 4: 2})

# Fit a model to get feature importances
model = XGBClassifier(random_state=42)
model.fit(X, y_mapped)

# Create a DataFrame of feature importances
importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': model.feature_importances_
}).sort_values(by='Importance', ascending=False)

# Show top 3
top_3 = importances.head(3)
print(top_3)

# Plot the top 3 important features
plt.figure(figsize=(8, 5))
sns.barplot(data=top_3, x='Importance', y='Feature', palette='viridis')
plt.title('Top 3 Important Features Affecting Employee Performance')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.tight_layout()
plt.show()
```

### Top 3 Important Factors Influencing Employee Performance



The top 3 features that significantly contribute to predicting employee performance, based on XGBoost feature importance, are:

1. EmpEnvironmentSatisfaction
2. YearsWithCurrManager
3. EmpJobSatisfaction

These insights can help HR teams focus on areas that have the most impact on performance.

To identify the most significant factors impacting employee performance, we evaluated both correlation analysis and model-driven feature importance. While correlation helped highlight linear relationships between features and performance ratings, we prioritized the results from the XGBoost classifier’s feature importance, which better captures non-linear and interaction effects crucial for predictive modeling.

The top three features influencing employee performance, as identified by the model, are:

EmpEnvironmentSatisfaction – Indicates how satisfied employees are with their working environment.

YearsSinceLastPromotion – Reflects career progression opportunities within the organization.

EmpLastSalaryHikePercent – Represents the perceived financial recognition of employees.

These insights suggest that environmental satisfaction, career advancement, and compensation growth play pivotal roles in employee performance.
